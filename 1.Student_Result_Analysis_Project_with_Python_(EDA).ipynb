# **Students Exam Scores**


In this project, we will understand how the student's result is influenced by the parent's education, Gender, ethnicity, lunch and test preparation. This dataset includes scores from three test scores of students at a  public school and different personal and socio-economic factors that may have interaction effects upon them.


First, we have to import necessary libraries that help us in exploratory data analysis on the data set students' exam scores.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Expanded_data_with_more_features.csv")
print(df.head)
print("Insight: the is 30641 rows and 15 column in the dataset")
```

# **Drop Unamed Column**

```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head)
print("Insight:Unnamed column is deleted,the total column is 14")
```

# **Gender Distribution**

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
print("Insight: From the above chart we have analyzed that The number of females is greater than the number of males. ")
```

# **Ethnic Group Distrubution**

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="EthnicGroup")
ax.bar_label(ax.containers[0])
plt.show()
print("Insight: From the above chart we have analyzed that The Group C EthnicGroup is greater than others EthnicGroup. ")
```

# **Group By Data On The Basis Of Parents Education,Parents Marital Status**
**the impact of parent education & Parents Marital Status **

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.title("Relationship between Parent's Education and Student's Score ")
plt.show()
print("Insight: From the above Chart we have concluded that the education of parents have a significant impact on students score.")
```

```python
gb2=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb2)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb2,annot=True)
plt.title("Relationship between Parent's Marital Status and Student's Score ")
plt.show()
print("Insight: From the above Chart we have concluded that the Marital status of parents have no impact or negligible impact on students score.")
```

# **Distribution Of Ethnic Group**

```python
print(df['EthnicGroup'].unique())
```

```python
from pandas.core.window.rolling import GroupbyIndexer
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()

mylist=[groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
l=['groupA','groupB','groupC','groupD','groupE',]
plt.pie(mylist,labels=l,autopct="%1.2f%%")
plt.title("Distribution Of Groups")
plt.show()
```
